```python
# all of this code is from youtube.com/watch?v=ok2s1vV9XW0
# this was mostly for learning/understanding purposes - I followed the tutorial which had code provided step-by-step


import pandas as pd
from sklearn.datasets import load_digits
digits = load_digits()
dir(digits)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
plt.gray()

# practice with displaying the data.
for i in range(4):
    plt.matshow(digits.images[i])

df = pd.DataFrame(digits.data)
df.head()
```

```python
df['target'] = digits.target
df[0:12]
```

```python
X = df.drop('target',axis='columns')
y = df.target


# train test split the data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
# using the random forest classifier, finding accuracy
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=20)
#training the model
model.fit(X_train, y_train)

# finding the accuracy using score.
model.score(X_test, y_test)

# prediction
y_predicted = model.predict(X_test)
```

```python
# practice creating a confusion matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_predicted)
cm
```

```python
#  using seaborn to plot and display the confusion matrix
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sn
plt.figure(figsize=(10,7))
sn.heatmap(cm, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
